```python
# setting up environment
import pandas as pd
import numpy as np
import os
import math

from dotenv import load_dotenv

load_dotenv()
```

```python
dataset = pd.read_csv('../datasets/framingham.csv')

dataset.head()
```

```python
# first of all, we will drop education columns, since it has no direct relation with coronel heart disease
def dropDataframeColumn(df, col):
  try:
    df = df.drop([col], axis=1)
    print("Column %s Dropped"%(col))
  except:
    print("Column dont exist or already deleted")
  return df

dataset = dropDataframeColumn(dataset,'education')
dataset.head()
```

```python
dataset.describe()
```

```python
# encoding categorical column to bolean (colomn sex and is_smoking)
def encodeColumnToBool(df, col, values):
  try:
    df[col] = df[col].replace(values,[0,1])
  except:
    print('invalid column to encode, not a str')
  return df

dataset = encodeColumnToBool(dataset, 'sex', ['F', 'M'])
dataset = encodeColumnToBool(dataset, 'is_smoking', ['NO', 'YES'])
dataset.head()
```

```python
# convert interger 0-1 to bolean
def convertColomnToInt(df, col):
    try:
        df[col] = dataset[col].astype(int)
    except:
        print('cant convert colomn to int')
    return df

dataset = convertColomnToInt(dataset,'BPMeds')
dataset.head()
```

```python
# check missing value

dataset.isnull().sum()
```

```python
# helper function to fix null data

def GenerateDataframeWhereColumnIsNull(df, col, query=''):
  df = pd.DataFrame(df[df[col].isnull()])
  if query == '':
    return df
  return df.query(query)

def GenerateDataframeWhereColumnIsNotNull(df, col, query=''):
  df = pd.DataFrame(df[df[col].notnull()])
  if query == '':
    return df
  return df.query(query)

def generateGroupedBy(df, col, target, mode="avg"):
  if mode == 'mode':
    return df.groupby([col], as_index=False)[target].agg(pd.Series.mode)
  return df.groupby([col], as_index=False)[target].mean()
```

```python
# next we will fix cigsPerDay, as we can see, null value only apprear in is smoking record
# we have to collect average of cigsPerDay of non null smoker

print("records which have null cigs per day")
print(GenerateDataframeWhereColumnIsNull(dataset, 'cigsPerDay').head())

nonNullCigsPerDaySmoker = GenerateDataframeWhereColumnIsNotNull(dataset, 'cigsPerDay', 'is_smoking == 1')
print("records which is smoking and cigs per day non null")
print(nonNullCigsPerDaySmoker.head())

avgCigsPerDay = math.floor(nonNullCigsPerDaySmoker['cigsPerDay'].mean())
print('average cigs per day smoker : %s' %(avgCigsPerDay))

# fill the smoker avg cigs per day to null cigs per day
dataset["cigsPerDay"].fillna(avgCigsPerDay, inplace = True)
```

```python
# next we will fix null bpmeds column
# bpmeds is blood pressure medicine consumtion
# we will replace null data with mode of record having same age

def generateGroupedBy(df, col, target, mode="avg"):
  if mode == 'mode':
    df = df.groupby([col])[target].agg(pd.Series.mode)
    return df
  return df.groupby([col], as_index=False)[target].mean()

# first we create map with key age and value BPMeds mode
mapAgeToBPMeds = (generateGroupedBy(dataset, 'age', 'BPMeds', 'mode'))
mapAgeToBPMeds

# since we got 0 in BPMeds in all ages, we can just replace missing BPMeds with 0
dataset['BPMeds'] = dataset['BPMeds'].fillna(0)
dataset.describe()
```

```python
# next we will fix null totchol column
# we will replace null data with avg of record having same age

# first we create map with key age and value totchol mean
mapAgeToTotchol = (generateGroupedBy(dataset, 'age', 'totChol'))

# then we map our dataset to grouped totchol by age
dataset['totChol'] = dataset['age'].map(mapAgeToTotchol.set_index('age')['totChol']).fillna(dataset['totChol'])
dataset.describe()
```

```python
# next we will fix null BMI column
# we will replace null data with avg of record having same age

# first we create map with key age and value bmi mean
mapAgeToTotchol = (generateGroupedBy(dataset, 'age', 'BMI'))

# then we map our dataset to grouped bmi by age
dataset['BMI'] = dataset['age'].map(mapAgeToTotchol.set_index('age')['BMI']).fillna(dataset['BMI'])
dataset.describe()
```

```python
# next we will fix null heartRate column
# we will replace null data with avg of record having same age

# first we create map with key age and value heartRate mean
mapAgeToTotchol = (generateGroupedBy(dataset, 'age', 'heartRate'))

# then we map our dataset to grouped heartRate by age
dataset['heartRate'] = dataset['age'].map(mapAgeToTotchol.set_index('age')['heartRate']).fillna(dataset['heartRate'])
dataset.describe()
```

```python
# next we will fix null glucose column
# we will replace null data with avg of record having same age

# first we create map with key age and value glucose mean
mapAgeToTotchol = (generateGroupedBy(dataset, 'age', 'glucose'))

# then we map our dataset to grouped glucose by age
dataset['glucose'] = dataset['age'].map(mapAgeToTotchol.set_index('age')['glucose']).fillna(dataset['glucose'])
dataset.describe()
```

```python
# double check for missing / null values
dataset.isnull().sum()
```

```python
# get number of positive cases, save balance testing rate

dataset['TenYearCHD'].value_counts()
```

```python
# divide datasets into testing and training datasets
# TESTINGRATE, ratio for testing data
TESTINGRATE = float(os.getenv('TESTINGRATE'))

# before we split the dataset, we shuffle the rows to randomize it
dataset = dataset.sample(frac=1).reset_index(drop=True)

testing_df = dataset.iloc[:int(len(dataset) * TESTINGRATE), :]
# dataset.iloc[:int(len(dataset) * TESTINGRATE), :]

training_df = dataset.iloc[int(len(dataset) * TESTINGRATE):, :]
# dataset.iloc[int(len(dataset) * TESTINGRATE):, :]

print('shape train dataset : {0}'.format(training_df.shape))
print('shape testing dataset : {0}'.format(testing_df.shape))
```

```python
# before we save the dataset, we shuffle the rows to randomize it
testing_df = testing_df.sample(frac=1).reset_index(drop=True)
training_df = training_df.sample(frac=1).reset_index(drop=True)
```

```python
# save training and testing data into csv

training_df.to_csv('../datasets/training.csv',  index=False)
testing_df.to_csv('../datasets/testing.csv',  index=False)
```
