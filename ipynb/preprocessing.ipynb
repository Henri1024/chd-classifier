```python
# setting up environment
import pandas as pd
import numpy as np
import os

from dotenv import load_dotenv

load_dotenv()
```

```python
dataset = pd.read_csv(os.getcwd()+'/datasets/dataset.csv')

dataset.head()
```

```python
dataset.describe()
```

```python
# encoding categorical column to bolean (colomn sex and is_smoking)
def encodeColumnToBool(df, col, values):
  try:
    df[col] = df[col].replace(values,[0,1])
  except:
    print('invalid column to encode, not a str')
  return df

dataset = encodeColumnToBool(dataset, 'sex', ['F', 'M'])
dataset = encodeColumnToBool(dataset, 'is_smoking', ['NO', 'YES'])
dataset.head()
```

```python
# convert interger 0-1 to bolean
def convertColomnToInt(df, col):
    try:
        df[col] = dataset[col].astype(int)
    except:
        print('cant convert colomn to int')
    return df

dataset = convertColomnToInt(dataset,'BPMeds')
dataset.head()
```

```python
# check missing value

dataset.isnull().sum()
```

```python
# divide datasets into testing and training datasets
# TESTINGRATE, ratio for testing data
TESTINGRATE = float(os.getenv('TESTINGRATE'))

# before we split the dataset, we shuffle the rows to randomize it
dataset = dataset.sample(frac=1).reset_index(drop=True)

training_df = dataset.iloc[int(len(dataset) * TESTINGRATE):, :]
testing_df = dataset.iloc[:int(len(dataset) * TESTINGRATE), :]

print('shape train dataset : {0}'.format(training_df.shape))
print('shape testing dataset : {0}'.format(testing_df.shape))
```

```python
# save training and testing data into csv

training_df.to_csv(os.getcwd()+'/datasets/training.csv',  index=False)
testing_df.to_csv(os.getcwd()+'/datasets/testing.csv',  index=False)
```
